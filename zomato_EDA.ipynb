```python
data.drop_duplicates(inplace=True)
data
```

```python
print("Total Restaurants:", data.shape[0])
```

```python
top_areas = data['area'].value_counts().head(10)
print(top_areas)
```

```python
# 1. Restaurants by Area
top_areas.plot(kind='bar')

plt.title("Top 10 Areas by Restaurant Count")
plt.xlabel("Area")
plt.ylabel("Number of Restaurants")
plt.xticks(rotation=45)
plt.show()
```

```python
# 2. Online Order Availability
data['online_order'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title("Online Order Availability")
plt.ylabel("")
plt.show()
```

```python
# 3. Table Booking Availability
data['table booking'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title("Table Booking Availability")
plt.ylabel("")
plt.show()
```

```python
# 4. Rating Distribution
plt.hist(data['rate (out of 5)'], bins=10)

plt.title("Restaurant Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

```python
# 5. Cost Distribution
plt.hist(data['avg cost (two people)'], bins=20)

plt.title("Average Cost for Two")
plt.xlabel("Cost")
plt.ylabel("Restaurants")
plt.show()
```

```python
# 1. Which Cities/Areas Have Most Restaurants?
data['area'].value_counts().head(10)
```

```python
# 2. Online Delivery vs Ratings
data.groupby('online_order')['rate (out of 5)'].mean()
```

```python
# 3. Most Popular Cuisines
data['cuisines type'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Cuisines")
plt.show()
```
